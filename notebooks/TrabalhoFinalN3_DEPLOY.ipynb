```python
# "Desempacotar" o Modelo Final (Random Forest)

import joblib

modelo = joblib.load("modelo_final.pkl")
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder

# carregador dados fake
df_fake = pd.read_csv("fake_dataset.csv")

# mesmsas colunas do treino
feature_cols = [
    'alpha', 'delta',
    'u', 'g', 'r', 'i', 'z',
    'run_ID', 'rerun_ID', 'cam_col', 'field_ID', 'spec_obj_ID',
    'redshift', 'plate', 'MJD', 'fiber_ID'
]


X_fake = df_fake[feature_cols]

# PREVER
y_pred = modelo.predict(X_fake)

# 5) refazer o LabelEncoder para voltar os nomes
le = LabelEncoder()
le.classes_ = np.array(['GALAXY', 'QSO', 'STAR'])

# 6) trocar numero da classse pelo mnomnome
y_pred_names = le.inverse_transform(y_pred)


# imprimir as respostas
print("===== PREVISÕES DO MODELO =====")
for i, p in enumerate(y_pred_names):
    print(f"Linha {i}: classe_predita = {p}")
```
